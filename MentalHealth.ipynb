```python
def print_number_of_trainable_model_parameters(model):
    trainable_model_params = 0
    all_model_params = 0
    for _, param in model.named_parameters():
        all_model_params += param.numel()
        if param.requires_grad:
            trainable_model_params += param.numel()
    return f"\ntrainable model parameters: {trainable_model_params}\nall model parameters: {all_model_params}\npercentage of trainable model parameters: {100 * trainable_model_params / all_model_params:.2f}%"
   
```

```python
from peft import LoraConfig, get_peft_model, TaskType

lora_config = LoraConfig(
    r=32, # Rank
    lora_alpha=32,
    target_modules=["q", "v"],
    lora_dropout=0.05,
    bias="none",
    task_type=TaskType.SEQ_2_SEQ_LM # FLAN-T5
)
```

```python
peft_model = get_peft_model(original_model, 
                            lora_config)
print(print_number_of_trainable_model_parameters(peft_model))
```

```python
def build_dataset(model_name, dataset, input_min_text_length, input_max_text_length):
    # Filter the dataset based on text length criteria
    dataset = dataset[(dataset['questionText'].str.len() > input_min_text_length) & 
                      (dataset['answerText'].str.len() <= input_max_text_length)]
    
    # Prepare tokenizer
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    
    def tokenize(sample):
        # Wrap each dialogue with the instruction
        prompt = f"Help them in their question in a therapy session. {sample['questionText']} Solution:"
        sample['answerText'] = tokenizer.encode(prompt, return_tensors='pt').tolist()[0]
        
        # This must be called "query", which is a requirement of our PPO library
        sample['query'] = tokenizer.decode(sample['answerText'])
        return sample
    
    # Tokenize each dialogue
    dataset = dataset.apply(tokenize, axis=1)
    
    # Split the dataset into train and test parts
    dataset_splits = dataset.sample(frac=0.8, random_state=42), dataset.drop(dataset.sample(frac=0.8, random_state=42).index)
    return dataset_splits

train_df = build_dataset(model_name=model_name,
                        dataset=train_df,
                        input_min_text_length=200,
                        input_max_text_length=1000)

```

```python
test_df = build_dataset(model_name=model_name,
                        dataset=test_df,
                        input_min_text_length=200,
                        input_max_text_length=1000)
print(test_df)
```

```python
validation_df = build_dataset(model_name=model_name,
                        dataset=validation_df,
                        input_min_text_length=200,
                        input_max_text_length=1000)
print(validation_df)
```

```python
def tokenize_function(example):
    start_prompt = 'Help them in their question in a therapy session.\n\n'
    end_prompt = '\n\nSolution: '
    
    # Handle potential non-string values
    if not isinstance(example["questionText"], str):
        example["questionText"] = ""
    if not isinstance(example["answerText"], str):
        example["answerText"] = ""
        
    prompt = start_prompt + example["questionText"] + end_prompt
    
    tokenized_instructions = tokenizer(prompt, padding="max_length", truncation=True)
    tokenized_responses = tokenizer(example["answerText"], padding="max_length", truncation=True)
    
    example['questionText'] = tokenized_instructions['input_ids']
    example['answerText'] = tokenized_responses['input_ids']
    
    return example

# Apply tokenize_function


# Define collator
def data_collator(features):
    input_ids = [feature['questionText'] for feature in features]
    labels = [feature['answerText'] for feature in features]
    
    return {
        'input_ids': torch.tensor(input_ids, dtype=torch.long),
        'labels': torch.tensor(labels, dtype=torch.long)
    }

# Training arguments
output_dir = f'./peft-model-{str(int(time.time()))}'
peft_training_args = TrainingArguments(
    output_dir=output_dir,
    auto_find_batch_size=True,
    learning_rate=1e-3,  # Higher learning rate than full fine-tuning.
    num_train_epochs=1,
    logging_steps=1,
    max_steps=1,
    remove_unused_columns=False
)

# Initialize Trainer
peft_trainer = Trainer(
    model=peft_model,  # Ensure peft_model is properly defined
    args=peft_training_args,
    train_dataset=train_df,
    eval_dataset=validation_df,
    data_collator=data_collator
)

peft_trainer.train()
```
